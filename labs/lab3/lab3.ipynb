```julia
using DifferentialEquations, Plots;
```

```julia
function two_armies(u, p, t)
    x, y = u
    a, b, c, h = p
    dx = -a*x - b*y + sin(t + 5) + 1
    dy = -c*x - h*y + cos(t + 5) + 1
    return [dx, dy]
end
```

```julia
u0 = [24000, 54000]
p = [0.4, 0.64, 0.77, 0.3]
tspan = (0,1)
```

```julia
problem = ODEProblem(two_armies, u0, tspan, p)
solution = solve(problem, Tsit5())

plot(solution, title = "Модель боевых действий для двух армий", label = ["Армия X" "Армия Y"], xaxis = "Время", yaxis = "Численность армий")
```

```julia
function one_army(u, p, t)
    x, y = u
    a, b, c, h = p
    dx = -a*x - b*y + sin(2*t) + 2
    dy = -c*x*y - h*y + cos(t) + 1
    return [dx, dy]
end
```

```julia
u0 = [24000, 54000]
p = [0.35, 0.67, 0.77, 0.45]
tspan = (0, 0.01)
```

```julia
problem = ODEProblem(one_army, u0, tspan, p)
solution = solve(problem, Tsit5())

plot(solution, title = "Модель боевых действий для армии \nи партизанских отрядов", label = ["Армия X" "Армия Y"], xaxis = "Время", yaxis = "Численность армий")
```
