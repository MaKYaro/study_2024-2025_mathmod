Построить фазовый портрет гармонического осциллятора и решение уравнения
гармонического осциллятора для следующих случаев:

1. Колебания гармонического осциллятора без затуханий и без действий внешней
силы
 $$\ddot{x} + 10x = 0,$$

2. Колебания гармонического осциллятора c затуханием и без действий внешней силы 
  
  $$\ddot x + 1.5 \dot x + 3 x = 0,$$

3. Колебания гармонического осциллятора c затуханием и под действием внешней силы 
   
   $$\ddot x + 0.6 \dot x + 13 x = cos(1.5t).$$
На интервале $t \in [0; 62]$ (шаг 0.05) с начальными условиями $x_0 = 0.8, \,\, y_0 = -1$

```julia
using DifferentialEquations, Plots;

# Начальные условия
tspan = (0,62)
u0 = [0.8, -1]
p1 = [0, 10]
p2 = [1.5, 3]
p3 = [0.6, 1]
f(t) = cos(1.5*t)

# Задание функции
function f1(u, p, t)
    x, y = u
    g, w = p
    dx = y
    dy = -g .*y - w^2 .*x
    return [dx, dy]
end

# Постановка проблемы и ее решение
problem1 = ODEProblem(f1, u0, tspan, p1)
sol1 = solve(problem1, Tsit5(), saveat = 0.05)

```

```julia
plot(sol1, label = ["x" "y"])
```

```julia
plot(sol1, idxs = (1, 2), xaxis = "x", yaxis = "y", label = "Зависимость x от y")
```

```julia
problem2 = ODEProblem(f1, u0, tspan, p2)
sol2 = solve(problem2, Tsit5(), saveat = 0.05)
```

```julia
plot(sol2, label = ["x" "y"])
```

```julia
plot(sol2, idxs = (1, 2), xaxis = "x", yaxis = "y", label = "Зависимость x от y")
```

```julia
function f2(u, p, t)
    x, y = u
    g, w = p
    dx = y
    dy = -g .*y - w^2 .*x .+f(t)
    return [dx, dy]
end
```

```julia
problem3 = ODEProblem(f2, u0, tspan, p3)
sol3 = solve(problem3, Tsit5(), saveat = 0.05)
```

```julia
plot(sol3, label = ["x" "y"])
```

```julia
plot(sol3, idxs = (1, 2), xaxis = "x", yaxis = "y", label = "Зависимость x от y")
```
