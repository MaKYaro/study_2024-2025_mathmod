```julia
#  
using DifferentialEquations, Plots;

#   ,   -
function DE(u, p, t)
    x, y = u
    a, b, c, d = p
    dx = a*x - b*x*y
    dy = -c*y + d*x*y
    return [dx, dy]
end

#  
u0 = [7,12]
p = [-0.45, -0.046, -0.47, -0.048]
tspan = (0.0, 50.0)
problem1 = ODEProblem(DE, u0, tspan, p)
solution1 = solve(problem1, Tsit5())

#     
plot(solution1, title = " -", xaxis = "", yaxis = " ", label = ["" ""], c = ["green" "purple"], box =:on)
```

```julia
plot(solution1, idxs = (1, 2), yaxis = "", xaxis = "", label = " ")
```
